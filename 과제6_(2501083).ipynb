```python
#39
def print_odd_numbers(n):
    for i in range(n+1):
        if i % 2 != 0:
            print(i, end=' ')

num = int(input("양수(숫자) 입력: "))
print_odd_numbers(num)
```

```python
#40
def print_multiple_of_3(n):
    if n % 3 == 0:
        print(f"{n}은/는 3의 배수")
    else:
        print(f"{n}은/는 3의 배수x")

num = int(input("숫자 입력: "))
print_multiple_of_3(num)
```

```python
#41 
def find_max_min(a, b, c, d):
    numbers = [a, b, c, d]
    return max(numbers), min(numbers)
n1, n2, n3, n4 = map(int, input("숫자 4개 입력 (공백 구분): ").split())
max_val, min_val = find_max_min(n1, n2, n3, n4)
print(f"최댓값: {max_val}, 최솟값: {min_val}")
```

```python
#43
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result

num = int(input("정수 n입력: "))

if num<0:
    print("0보다 크거나 같은 정수 입력")
else:
    print(f"{num}! = {factorial(num)}")
```

```python
#44
def sum_products_over_30(limit_i, limit_j):
    total_sum = 0
    for x in range(1, limit_i + 1):
        for y in range(1, limit_j + 1):
           product = x*y
           if product >= 30:
                total_sum += product
    return total_sum

i, j = map(int, input("2이상 9이하 양수 2개 입력: ").split())
result = sum_products_over_30(i, j)
print(f"곱이 30이상인 경우의 총 합: {result}")
```

```python
#45
def calculate_list_sum(numbers_list):
    total = 0
    for num in numbers_list:
        total += num
    print(f"리스트 값의 (누적) 합: {total}")

a = [1, 2, 3, 4, 5]

calculate_list_sum(a)
```
